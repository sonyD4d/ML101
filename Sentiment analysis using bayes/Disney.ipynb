```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import re

snow = SnowballStemmer(language='english')

def pre_process(x):
    x = re.sub('[^a-zA-Z]', ' ', x)
    x = x.lower()
    x = word_tokenize(x)
    x = [snow.stem(word) for word in x if not word in set(stopwords.words('english'))]
    x = ' '.join(x)
    return x
```

```python
data = pd.read_csv('./pre.csv')
data = data.drop(columns=['Unnamed: 0'])
data
```

```python
x = list(data['Review_Text'])
y = list(data['Rating'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
bow = pd.DataFrame( cv.fit_transform(x).toarray() , columns=cv.get_feature_names())
```

```python
bow.shape
```

```python
voc = bow.shape[1]
```

```python
bow['_Class']=y
bow
```

```python
voc = bow.shape[1]
```

```python
rating = list(set(y))
fc_probability = dict()
c_probability = dict()
for i in rating:
    outer = bow[bow['_Class'] == i]
    c_probability[i] = outer.shape[0]/len(y)
    outer.drop(columns=['_Class'])
    add = outer.sum()
    dr = sum(list(add)) + voc
    for j in list(bow.columns[0:-1]):
        p = (add[j] + 1)/dr
        fc_probability[str(j)+"/"+str(i)] = p
```

```python
c_probability
```

```python
fc_probability
```

```python
def predict(s):
    best_class = -1
    best_prob = -1
    for i in rating:
        prob = c_probability[i]
        for word in word_tokenize(s):
            if(len(word)>1):
                prob *= fc_probability[word+"/"+str(i)]
        if(prob>best_prob):
            best_prob = prob
            best_class = i
    return best_class
```

```python
count = 0
for index, row in data.iterrows():
    if(row["Rating"] == predict(row["Review_Text"])):
            count+=1
print("Accuracy: "+ str((count/len(y))*100))
```
